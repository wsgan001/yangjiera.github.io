# Text Processing & Retrieval Pipeline

This Python notebook demonstrates the basic text processing pipeline using "How I Met Your Mother" dataset. 

Following steps will be covered:
- Loading data
- Tokenization
- Stop words removal
- Stemming & lemmatization


- Creating term-document matrix (TF-IDF)
- Query with Vector Space Model
- Query with BM25
- Query with BM25 in Elasticsearch

## Load data

```python
# Dataset preview

from IPython.display import Image
Image('ScreenShot.png')
```

```python
# Read in dataset into a python list, i.e. "corpus"
# Each element of the list corresponds to text in an episode

from collections import defaultdict
import csv

# Read in each episode as a list of sentences
episodes = defaultdict(list)
with open("sentences.csv", "r") as sentences_file:
    reader = csv.reader(sentences_file, delimiter=',')
    reader.next()
    for row in reader:
        episodes[row[1]].append(row[4])
        
# Join sentences in the same episode to a long string
for episode_id, text in episodes.iteritems():
    episodes[episode_id] = "\n".join(text)

# Make a corpus as a list of episodes
corpus = []
for id, episode in sorted(episodes.iteritems(), key=lambda t: int(t[0])):
    corpus.append(episode)
    
print "there are in total {} episides.".format(len(corpus))
```

```python
# See how the first element (episode) of the corpus looks like.

print corpus[0]
```

# Tokenization

```python
# remove punctuation and tokenize the corpus

from nltk import word_tokenize
import re

corpus_tokenized = []

for episode in corpus:
    # remove punctuation (using regular expression)
    letters_only = re.sub("[^a-zA-Z]", " ", episode) 
    
    # tokenization using nltk package
    tokens = word_tokenize(letters_only.lower().decode('utf-8'))
    corpus_tokenized.append(tokens)
```

```python
print corpus_tokenized[0]
```

```python
# Print the top-20 most frequent words

from nltk import FreqDist
def print_top_words(n, corpus):
    corpus_merged = []
    for episode in corpus:
        corpus_merged.extend(episode)
        
    print "there are in total {} words, and {} unique words".format(len(corpus_merged), len(set(corpus_merged)))    
    fdist = FreqDist(corpus_merged)
    top_word_count = fdist.most_common(n)
    i = 0
    print "top words:"
    for top_word in top_word_count:
        print i, top_word[0], top_word[1]
        i += 1

print_top_words(20, corpus_tokenized)
```

## Stop words removal

```python
# Load in English stop words

from nltk.corpus import stopwords # Import the stop word list

print stopwords.words("english")
stopwords_english = set(stopwords.words("english"))
```

```python
# Remove stop words in all espisodes of the corpus

corpus_stopwords_removed = []
for episode_tokenized in corpus_tokenized:
    words = [w for w in episode_tokenized if not w in stopwords_english]
    corpus_stopwords_removed.append(words)
```

```python
# Look at the top-word list now. It makes much more sense!

print_top_words(20, corpus_stopwords_removed)
```

## Stemming and lemmatization

### Stemming

```python
# We use again nltk package for stemming and lemmatization

from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()

# Two examples
print porter_stemmer.stem('going')
print porter_stemmer.stem('presumably')
```

```python
# Stem the entire corpus

corpus_stemmed = []
for episode_stopwords_removed in corpus_stopwords_removed:
    words = [porter_stemmer.stem(w) for w in episode_stopwords_removed]
    corpus_stemmed.append(words)
```

```python
# The new top-word list. "go" got boosted. "lili" is not a good example...

print_top_words(20, corpus_stemmed)
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Example
print wordnet_lemmatizer.lemmatize('dogs')
print wordnet_lemmatizer.lemmatize('churches')
```

```python
# Lemmatize the entire corpus

corpus_lemmatized = []
for episode_stemmed in corpus_stemmed:
    words = [wordnet_lemmatizer.lemmatize(w) for w in episode_stemmed]
    corpus_lemmatized.append(words)
```

```python
# Show the frquency distribution of the top-5000 words

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def graph(n, corpus):
    corpus_merged = []
    for episode in corpus:
        corpus_merged.extend(episode)
    fdist = FreqDist(tokens)
    top_word_count = fdist.most_common(n)
    
    data_y = [top_word[1] for top_word in top_word_count]
    data_x = np.arange(1,len(data_y)+1)
    
    plt.plot(data_x, data_y, linewidth=2)
    plt.xlabel('word index')
    plt.ylabel('frequency')
    plt.xscale('log')
    plt.yscale('log')
    
    
graph(5000, corpus_lemmatized)
```

## Creating the term-document matrix

```python
# Join words in the same episode as a long string

corpus_result = []
for episode_lemmatized in corpus_lemmatized:
    corpus_result.append(' '.join(episode_lemmatized))
```

```python
# Now an episode looks like this. Compare to the original text!

print corpus_result[0]
```

```python
# To index the corpus into a term-document matrix, we use a package from sklearn

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# vectorizer = CountVectorizer()
vectorizer = TfidfVectorizer()
corpus_indexed = vectorizer.fit_transform(corpus_result)
feature_names = vectorizer.get_feature_names()
```

```python
# How does the term-document matrix look like?

# print corpus_indexed.toarray().shape
# print feature_names
print type(corpus_indexed)
print corpus_indexed
```

## Query with Vector Space Model

```python
# The vectorizer can actually be used to transform any document

query = '''Narrator: Kids, I'm going to tell you an incredible story. The story of how I met your mother
Son: Are we being punished for something?
Narrator: No
Daughter: Yeah, is this going to take a while?
Narrator: Yes. (Kids are annoyed) Twenty-five years ago, before I was dad, I had this whole other life.
(Music Plays, Title "How I Met Your Mother" appears)
'''

# Remove punctuation then tokenize
letters_only = re.sub("[^a-zA-Z]", " ", query)     
query_tokened = word_tokenize(letters_only.lower().decode('utf-8'))
# Stop words removal
query_stopwords_removed = [w for w in query_tokened if not w in stopwords_english]
# Stemming and Lemmatization
query_stemmed = [porter_stemmer.stem(w) for w in query_stopwords_removed]
query_lemmatized = [wordnet_lemmatizer.lemmatize(w) for w in query_stemmed]

# Transform the query using TFIDF vectorizer
query_indexed = vectorizer.transform([' '.join(query_lemmatized)])

for item in query_indexed.indices:
    print "Word: {}; TFIDF: {}".format(feature_names[item], query_indexed[0,item])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosMat = cosine_similarity(query_indexed,corpus_indexed)
print cosMat

# Find the top-5 most relevant episode
related_docs_indices = cosMat[0].argsort()[:-6:-1]

print '\n\nThe index numbers for top-5 results are:'
print '(Episode Number | Cosine Similarity)\n'
for item in related_docs_indices:
    print 'Episode', item, '|', cosMat[0][item]
```

## Query with BM25

Documents
    0. Human machine interface for lab abc computer applications
    1. A survey of user opinion of computer system response time
    2. The EPS user interface management system
    3. System and human system engineering testing of EPS
    4. Relation of user perceived response time to error measurement
    5. The generation of random binary unordered trees
    6. The intersection graph of paths in trees
    7. Graph IV Widths of trees and well quasi ordering
    8. Graph minors A survey
Query
    
    The intersection graph of paths in trees survey Graph

```python
from bm25 import *

fn_docs = 'mycorpus.txt'

# Initialize the BM25 model 
bm25 = BM25(fn_docs, delimiter=' ')
    
print '...Vocabulary...'
print bm25.Items()
print '...DocIDF...'
print bm25.DocIDF
print '...DocTF...'
print bm25.DocTF
print '...DocLen...'
print bm25.DocLen
print '...DocAvgLen...'
print bm25.DocAvgLen
```

```python
# BM25 relevance rank ========================
Query = 'The intersection graph of paths in trees survey Graph'

# Split the query into individual words 
Query = Query.split()

def retrieve_docs(scores):
    doc_rank = []
    for idx, x in enumerate(scores):
        doc_rank.append([idx, x])
    doc_rank.sort(key=lambda x: x[1], reverse=True)
    print '(Document Id | Relevance)'
    for dr in doc_rank:
        print dr[0], '|', dr[1]
    print '\n'
    
k1, b = 1.2, 0.75
scores = bm25.BM25Score(Query, k1, b)
retrieve_docs(scores)
```

## Query with BM25 in Elasticsearch

```python
# insert data into elasticsearch

from elasticsearch import Elasticsearch
es = Elasticsearch()

docs = [{
            "id": "0",
            "text":  "Human machine interface for lab abc computer applications",
        }, {
            "id": "1",
            "text":  "A survey of user opinion of computer system response time",
        }, {
            "id": "2",
            "text": "The EPS user interface management system",
        }, {
            "id": "3",
            "text":  "System and human system engineering testing of EPS",
        }, {
            "id": "4",
            "text":  "Relation of user perceived response time to error measurement",
        }, {
            "id": "5",
            "text":  "The generation of random binary unordered trees",
        }, {
            "id": "6",
            "text":  "The intersection graph of paths in trees",
        }, {
            "id": "7",
            "text":  "Graph IV Widths of trees and well quasi ordering",
        }, {
            "id": "8",
            "text":  "Graph minors A survey",
        }]

for i in range(9):
    # PUT
    res = es.index(
        index="bm25_test", doc_type='table', id=i + 1, body=docs[i])
    # GET
    res = es.get(index="bm25_test", doc_type='table', id=i + 1)
    print(res['_source'])

print '***************Index ends here***************'
```

```python
# Query elasticsearch

qry = {"query": 
           {"match": 
                {"text": "The intersection graph of paths in trees survey Graph"}
           }
      }

res = es.search(index="bm25_test", doc_type="table", body=qry)

for doc in res['hits']['hits']:
    print("%s | %s" % (doc['_source']['id'], doc['_source']['text']))
```
